```python
n=3
def get_data(n: int):
    FILE=f"n{n}-sorted.txt"
    data = []
    with open(FILE, "r") as f:
        for line in f:
            val = int(line.strip())
            data.append(val)
    return data
```

```python
import matplotlib.pyplot as plt 
plt.rc('text',usetex=True)

plt.rcParams["figure.figsize"] =(5, 3)
plt.rcParams.update({'font.size': 13})

data = {
    3: get_data(3),
    # 9: get_data(9),
    # 17:get_data(17),
    # 33: get_data(33),
}

for k in data.keys():
    plt.hist(data[k], bins=70, label=f"OptRand ($n={k}$)")
# plt.xlabel
plt.legend()
plt.xlabel("Beacon Latency (in ms)")
plt.ylabel("Frequency")
plt.show()
# plt.savefig(f"../../Overleaf/OptRand/Figures/Dist.pdf", bbox_inches='tight')
```
