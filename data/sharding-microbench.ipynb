```python
FILE="acc.txt"

data = {}

with open(FILE, "r") as f:
    for line in f:
        words = line.split()
        key = words[0].split("/")
        dkey = key[0]
        n = int(key[1])
        if dkey not in data.keys():
            data[dkey] = {}
        # print(words[4],words[5],words)
        if words[5] == "ms":
            data[dkey][n] = float(words[4])*(10**-3)
        elif words[5] == "us":
            data[dkey][n] = float(words[4])*(10**-6)
    # print(data)
```

```python
import matplotlib.pyplot as plt

plt.rc('text',usetex=True)

plt.rcParams["figure.figsize"] =(5, 3)
plt.rcParams.update({'font.size': 13})

for key in data:

    x = data[key].keys()
    y = [data[key][j]*1000 for j in x]
    key = key.replace("_", " ")
    plt.plot(x, y, label=f"{key}", linestyle='--', marker='*')

plt.legend()
# plt.xlim([0,70])

plt.xlabel("Number of Nodes ($n$)")
plt.ylabel("Time in $ms$")
# plt.show()
# plt.xticks(log_points)
plt.savefig(f"../../Overleaf/OptRand/Figures/Acc-Benchmarks.pdf", bbox_inches='tight')
```
